```python
import fastparquet
import pandas as pd
ipdata=pd.read_parquet('/stuff/explodedipdata')
```

```python
ipdata=ipdata.drop_duplicates(subset='A')
```

```python
goodguys=ipdata.loc[lambda ipdata:ipdata['benign'] > 0,:]
badguys=ipdata.loc[lambda ipdata:ipdata['benign']==0,:]
```

```python
goodguys['A'].value_counts()
```

```python
badguys['A'].value_counts()
```

```python
new=ipdata['A'].str.split(".",n=-1,expand=True)
```

```python
ipdata['1st Octet']=new[0]
ipdata['2nd Octet']=new[1]
ipdata['3rd Octet']=new[2]
ipdata['4th Octet']=new[3]
```

```python
ipdata.drop(columns=['A'],inplace=True)
```

```python
ipdata.to_parquet("octetdata")
```
