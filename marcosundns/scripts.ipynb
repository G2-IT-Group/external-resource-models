```python
import pandas as pd
import os
import json
from pandas.io.json import json_normalize


#convert json to pd.dataframe

dnsdata=pd.DataFrame()
directory=os.fsencode('/data/data/2019-06-07/enrichment/dns')


for file in os.listdir(directory):
    with open(os.fsdecode(directory) + '/' +os.fsdecode(file)) as foo:
        bar = json.load(foo)
    foobar=json_normalize(bar['data'])

    if len(os.fsdecode(file))<12:
        foobar.insert(2,'benign',True)
    else:
        foobar.insert(2,'benign',False)  
           
    dnsdata=dnsdata.append(foobar)
         
dnsdata=dnsdata.reset_index(drop=True)
dnsdata=dnsdata[['indicator','benign','A','AAAA','CAA','CNAME','DNAME','LOC','MX','RP','SIG','SSHFP','success']]



#show all data
pd.set_option('display.max_columns', None)  
pd.set_option('display.max_rows', None) 
pd.set_option('display.max_colwidth', -1) 
dnsdata.head(15)
```

```python
import pandas as pd
import os
import json
from pandas.io.json import json_normalize
from json import JSONDecodeError
moredata=pd.DataFrame()
baddata=[]

toplevel=os.fsencode('/data/data/')
for date in os.listdir(toplevel):
    nextlevel=os.fsencode(os.fsdecode(toplevel)+os.fsdecode(date)+'/enrichment/dns/')
    try:
        for file in os.listdir(nextlevel):
            try:
                with open(os.fsdecode(nextlevel)+os.fsdecode(file)) as data:
                    temp=json.load(data)
                    prettytemp=json_normalize(temp['data'])
                    moredata=pd.concat([moredata,prettytemp],ignore_index=True)
                    print(os.fsdecode(nextlevel)+os.fsdecode(file))
            except AttributeError:
                print('error at '+os.fsdecode(nextlevel)+os.fsdecode(file))
                baddata=baddata+[os.fsdecode(nextlevel)+os.fsdecode(file)]
            except JSONDecodeError:
                print('empty json file at '+os.fsdecode(nextlevel)+os.fsdecode(file))
                baddata=baddata+[os.fsdecode(nextlevel)+os.fsdecode(file)]
    except FileNotFoundError:
        print("no enrichment data found for " + os.fsdecode(date))
        baddata=baddata+[os.fsdecode(date)]
        
        
            
```

```python
#cleanup function for weird list string things

temp=dnsdata.copy()

def cleanup(dirtydata):
    if type(dirtydata) == list:
        cleandata=dirtydata[0].split('\n')
        return cleandata
    elif type(dirtydata) == str:
        return 'No Answer'
    elif type(dirtydata) == float:
        return float('nan')

temp['A']=temp['A'].apply(cleanup,'columns')
temp['AAAA']=temp['AAAA'].apply(cleanup,'columns')
temp['CAA']=temp['CAA'].apply(cleanup,'columns')
temp['CNAME']=temp['CNAME'].apply(cleanup,'columns')
temp['DNAME']=temp['DNAME'].apply(cleanup,'columns')
temp['LOC']=temp['LOC'].apply(cleanup,'columns')
temp['MX']=temp['MX'].apply(cleanup,'columns')
temp['RP']=temp['RP'].apply(cleanup,'columns')
temp['SIG']=temp['SIG'].apply(cleanup,'columns')
temp['SSHFP']=temp['SSHFP'].apply(cleanup,'columns')

temp.to_parquet('dnsdata.parquet')
```

```python
#translate to basic qualitative variables for regression

regresstemp=temp.copy()
def regresstranslate(raw):
    if type(raw) == list or raw == True:
        return 1
    else:
        return 0
    
regresstemp['benign']=regresstemp['benign'].apply(regresstranslate,'columns')
regresstemp['success']=regresstemp['success'].apply(regresstranslate,'columns')
regresstemp['AAAA']=regresstemp['AAAA'].apply(regresstranslate,'columns')
regresstemp['A']=regresstemp['A'].apply(regresstranslate,'columns')
regresstemp['AAAA']=regresstemp['AAAA'].apply(regresstranslate,'columns')
regresstemp['CAA']=regresstemp['CAA'].apply(regresstranslate,'columns')
regresstemp['A']=regresstemp['A'].apply(regresstranslate,'columns')
regresstemp['CNAME']=regresstemp['CNAME'].apply(regresstranslate,'columns')
regresstemp['DNAME']=regresstemp['DNAME'].apply(regresstranslate,'columns')
regresstemp['LOC']=regresstemp['LOC'].apply(regresstranslate,'columns')
regresstemp['MX']=regresstemp['MX'].apply(regresstranslate,'columns')
regresstemp['RP']=regresstemp['RP'].apply(regresstranslate,'columns')
regresstemp['SIG']=regresstemp['SIG'].apply(regresstranslate,'columns')
regresstemp['SSHFP']=regresstemp['SSHFP'].apply(regresstranslate,'columns')
regresstemp
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

logmodel=LogisticRegression()

X_train, X_test, y_train, y_test = train_test_split(regresstemp.loc[:,'A':],regresstemp['benign'])

logmodel.fit(X_train,y_train)
predictions=logmodel.predict(X_test)
print(classification_report(y_test,predictions))
```
