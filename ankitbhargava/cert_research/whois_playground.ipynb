```python
def parse_whois(x):
    singleRes=[]
    try:
        book=json.load(open(x))['data']
    except:
        return singleRes
    for whoisrec in book:
        try:
            if (type(whoisrec['whois']) == str ):
                tmp=re.findall('[\w\s]*:.*\n',whoisrec['whois'])
            elif (type(whoisrec['whois']) == dict):
                tmp=whoisrec['whois']
            else:
                print(type(whoisrec['whois']))
            
            singleRes.append(tmp)
        except Exception as e:
            continue
            
    return singleRes
            
results=[]
pool = mp.Pool(4)
results = pool.map(parse_whois, everyFile)
pool.close()
```

```python
def pullDataframe(val):
    alldfs=[]
    try:
        for x in val:
            if (type(x) == dict):
                try:
                    alldfs.append(pd.DataFrame(x,index=[0]))
                except:
                    continue
            elif (type(x) == list):
                todf={}
                for y in x:
                    todf[y[0:y.find(':')]]=y[y.find(':')+1:].strip()

                alldfs.append(pd.DataFrame(todf,index=[0]))
    except Exception as e:
        return None
    
    return alldfs
        
next_results=[]
pool = mp.Pool(4)
new_results = pool.map(pullDataframe, results)
pool.close()
```
