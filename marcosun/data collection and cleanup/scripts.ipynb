```python
import pandas as pd
import os
import json
from pandas.io.json import json_normalize
from json import JSONDecodeError
moredata=pd.DataFrame()
baddata=[]

toplevel=os.fsencode('/data/data/')
for date in os.listdir(toplevel):
    nextlevel=os.fsencode(os.fsdecode(toplevel)+os.fsdecode(date)+'/enrichment/dns/')
    try:
        for file in os.listdir(nextlevel):
            try:
                with open(os.fsdecode(nextlevel)+os.fsdecode(file)) as data:
                    temp=json.load(data)                      
                    prettytemp=json_normalize(temp['data'])
                    try:
                        if len(os.fsdecode(file))<15:
                            prettytemp.insert(2,'benign',True)
                        else:
                            prettytemp.insert(2,'benign',False)
                        moredata=pd.concat([moredata,prettytemp],ignore_index=True)
                        print(os.fsdecode(nextlevel)+os.fsdecode(file))
                    except IndexError:
                        print('No data found for '+os.fsdecode(nextlevel)+os.fsdecode(file))
                        baddata=baddata+[os.fsdecode(nextlevel)+os.fsdecode(file)]
            except AttributeError:
                print('error at '+os.fsdecode(nextlevel)+os.fsdecode(file))
                baddata=baddata+[os.fsdecode(nextlevel)+os.fsdecode(file)]
            except JSONDecodeError:
                print('empty json file at '+os.fsdecode(nextlevel)+os.fsdecode(file))
                baddata=baddata+[os.fsdecode(nextlevel)+os.fsdecode(file)]
    except FileNotFoundError:
        print("no enrichment data found for " + os.fsdecode(date))
        baddata=baddata+[os.fsdecode(date)]          
```

```python
#cleanup function for weird list string things
import fastparquet
temp=moredata.copy()
nxdata=pd.DataFrame()

def cleanup(dirtydata):
    if type(dirtydata) == list:
        cleandata=dirtydata[0].split('\n')
        return cleandata
    else:
        return dirtydata

temp=temp.applymap(cleanup)

nxdata=temp[~temp['A'].notnull()]
temp=temp[temp['A'].notnull()]


```

```python
pd.set_option('display.max_columns', None) 
pd.set_option('display.max_colwidth', -1)
dnsdata=temp.drop(['NXDOMAIN',
       'NXDOMAIN.indicator', 'NXDOMAIN.reported', 'NXDOMAIN.source',
       'NXDOMAIN.source_data.content', 'NXDOMAIN.source_data.created',
       'NXDOMAIN.source_data.description', 'NXDOMAIN.source_data.id',
       'NXDOMAIN.source_data.indicator', 'NXDOMAIN.source_data.pulse_key',
       'NXDOMAIN.source_data.slug', 'NXDOMAIN.source_data.title',
       'NXDOMAIN.source_data.type', 'NXDOMAIN.success', 'NXDOMAIN.type','error'],axis=1)

nxdata=nxdata.drop(['A', 'AAAA', 'CAA', 'CNAME', 'DNAME', 'LOC', 'MX'],axis=1)
```

```python
def tolist(data):
    if type(data)==str:
        return [data]
    else:
        return data

dnsdata=dnsdata.applymap(tolist)
```

```python
finaldata=dnsdata[['indicator','benign','A', 'AAAA', 'CAA', 'CNAME', 'DNAME', 'LOC', 'MX', 'RP', 'SIG', 'SSHFP',
         'success']]
finaldata.to_parquet('dnsdata')
```
