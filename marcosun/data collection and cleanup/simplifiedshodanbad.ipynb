```python
import pandas as pd
import json
import os
shodanframe=pd.DataFrame(columns={"indicator","ip_str","ports","country_code","org","isp"})
toplevel="/data/data/"
for dates in os.listdir(toplevel):
    try:
        print(dates)
        for datum in os.listdir("/data/data/"+dates+"/enrichment/shodan/"):
                if len(datum)>10:
                    with open("/data/data/"+dates+"/enrichment/shodan/"+datum) as data:
                        try:
                            stuff=json.load(data)
                            for i in range(len(stuff['data'])):
                                products=set()
                                if(stuff['data'][i]['success']==False):
                                    pass
                                else:
                                    try:
                                        for samples in range(len(stuff['data'][i]['data']['data'])):
                                            try:
                                                products.add(stuff['data'][i]['data']['data'][samples]['product'])
                                            except KeyError:
                                                pass
                                    except KeyError:
                                        for ports in stuff['data'][i]['data']['data']:
                                            try:
                                                products.add(ports['product'])
                                            except KeyError:
                                                pass
                                    tobeadded=pd.DataFrame({"indicator":[stuff['data'][i]["indicator"]],"ip_str":[stuff['data'][i]['data']["ip_str"]],"ports":[stuff['data'][i]['data']["ports"]],"products":[products],"country_code":[stuff['data'][i]['data']["country_code"]],"org":[stuff['data'][i]['data']["org"]],"isp":[stuff['data'][i]['data']["isp"]],"os":[stuff['data'][i]['data']['os']]})
                                    shodanframe=shodanframe.append(tobeadded,ignore_index=True)
                        except ValueError:
                            print("valueerror at "+dates+"/"+datum)
                        except TypeError:
                            print("typeerror at "+dates+"/"+datum)

                        
    except FileNotFoundError:
        pass
```
