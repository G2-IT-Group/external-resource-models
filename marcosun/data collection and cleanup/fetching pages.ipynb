```python
import pandas as pd
import cryptography
import os 
import json
toplevel="/data/data"
for dates in os.listdir(toplevel):
    print(dates)
    nextlevel=toplevel+"/"+dates+"/enrichment/fetch_page"
    for data in os.listdir(nextlevel):
        print(data)
        with open(nextlevel+"/"+data) as toload:
            file=json.load(toload)
```

```python
file['data'][0]["status_code"]
```

```python
key=file['data'][0]['encryption_key']
content=file['data'][0]["page_content"]
```

```python

from cryptography.fernet import Fernet
bkey=bytes(key,"utf-8")

fkey=Fernet(bkey)

bcontent=bytes(content,"utf-8")
dc=fkey.decrypt(bcontent)
```

```python
dc
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(dc,"html.parser")
```

```python
soup
for link in soup.find_all('a'):
    print(link.get('href'))
```
