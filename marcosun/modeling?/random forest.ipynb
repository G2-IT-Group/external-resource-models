```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
ipdata=pd.read_parquet('/stuff/data collection and cleanup/octetdata')
```

```python
ipdata
```

```python
def toint(cell):
    cell = int(cell)
    return cell
ipdata=ipdata.applymap(toint)
```

```python
goodguys=ipdata.loc[lambda ipdata:ipdata['benign'] > 0,:].reset_index(drop=True)
badguys=ipdata.loc[lambda ipdata:ipdata['benign']==0,:].reset_index(drop=True)

badguys=badguys.truncate(after=10000)
balancedip=goodguys.append(badguys)
balancedip=balancedip.reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(balancedip.loc[:,'ttl':'4th Octet'],balancedip['benign'])
```

```python
model=RandomForestClassifier()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```
