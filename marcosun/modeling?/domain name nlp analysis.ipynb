```python
import pandas as pd

data=pd.read_parquet('/stuff/dnsdata')
```

```python
domainnames=data[['indicator','benign']]
```

```python
def tostring(row):
    row['indicator']=row['indicator'][0]
    return row

domainnames=domainnames.apply(tostring,axis='columns')
```

```python
domainnamesunique=domainnames.drop_duplicates('indicator')
```

```python
domainnamesneatandunique=domainnamesunique.reset_index(drop=True)
```

```python
def pushitover(row):
    i=len(row)-1
    while row[i] is None:
        i=i-1
    row=row.shift(len(row)-1-i,fill_value='None')
    return row

def removewww(cell):
    if cell=="www":
        cell="None"
    return cell

test=domainnamesneatandunique['indicator'].str.split('.',n=-1,expand=True)
test2=test.apply(pushitover,axis="columns")
test2.insert(0,'Benign',domainnamesneatandunique['benign'])
test2=test2.applymap(removewww)
```

```python
test2sliced=test2.loc[:,['Benign',13,14,15,16,17]]
```
