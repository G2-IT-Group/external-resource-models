```python
def change_ln(input):
    if(type(input)==float):
        if(math.isnan(input)==True):
            return 0
    else:
        x = str(input)
        return len(x)
```

```python
def change_nan(input):
    if(type(input)==float):
        if(math.isnan(input)==True):
            return 0
    else:
        return input
        
```

```python
# counts number of html there are
```

```python
def count_html(input):
    if(type(input)==float):
        if(math.isnan(input)==True):
            return 0
    else:
        # gives a list
        x = str(input)
        html_match = re.findall('html',x)
        # returns the length of the list
        return len(html_match)
```

```python
# the following 2 are tricks to change 0 to a tring and back
```

```python
def change_zero_to_0(input):
    if(input==0):
        return "zero"
    else:
        return input
```

```python
def change_0_to_zero(input):
    if(input=="zero"):
        return 0
    else:
        return input
```

```python
# counts forward slashes
```

```python
def num_of_forward_slash(input):
    if(type(input)==float):
         if(math.isnan(input)==True):
            return 0
    else:
        x = str(input)
        slash_match = re.findall('\/',x)
        return len (slash_match)
```

```python
# counts the number of the word div
```

```python
def num_of_div(input):
    if(type(input)==float):
         if(math.isnan(input)==True):
            return 0
    else:
        x = str(input)
        slash_match = re.findall('\div',x)
        return len (slash_match)
```
